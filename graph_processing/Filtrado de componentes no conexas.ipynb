# Filtrado de componentes no conexas.

A continuación se generan con el LaNet-Vi distintos gráficos en svg y png procesando los grafos generados en esta [notebook](./CAIDA.ipynb).  
Como el Lanet-Vi solo funciona con grafos conexos lo primero que se hará es filtrar los grafos obtenidos dejando solo la componente conexa. Los grafos de entrada serán tomados de "./grafos_sn/" y su componente conexa mas grande será guardada con el mismo nombre en "./con_graphs/".

```python
import pandas as pd
import random
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
import re
for file in sorted(listdir("./grafos_sn/")):
    print("Procesando el archivo:" + file)
    G=nx.read_adjlist("./grafos_sn/"+file)
    #grafo=pandas.read_csv("./grafos_sn/"+file, delimiter=' ', names=["AS1", "AS2"], engine='python',dtype={'AS1':object,'AS2':object})
    #grafo_proc=grafo[grafo["AS1"].str.isdigit() & grafo["AS2"].str.isdigit()]
    #grafo_proc.to_csv("./grafos_sn/"+file, index=False, header=False, sep=" ")
    for graph in sorted(nx.connected_component_subgraphs(G), key = len, reverse=True):
        print("Se encontró un subgrafo conexo de tamaño: " + str(len(graph)))
    G=sorted(nx.connected_component_subgraphs(G), key = len, reverse=True)[0]
    #nx.write_adjlist(sorted(nx.connected_component_subgraphs(G), key = len, reverse=True)[0],"./con_graphs/"+file)
    G.remove_edges_from(G.selfloop_edges()) #Agregué esta línea para eliminar los self loops
    df=pd.DataFrame(G.edges())
    df.to_csv("./con_graphs/"+file, index=False, header=False, sep=" ")
    with open("./con_graphs/"+file,"r") as f:
        data = f.read()
        data = re.sub(r'#.*', "", data)
        data = re.sub(r'\n\n\n', "", data)
    with open("./con_graphs/"+file, "w") as f:
        f.write(data)
        f.close()
```

## Mapeo de número de nodo mayor a 65535

A partir de marzo del 2013 se detectó que hay números de nodo mayores a 65535. Como el LaNet-vi reserva memoria según el número de nodo de mayor orden, vamos a mapear los números de nodo mayores a 65535 a números consecutivos.

```python
import pandas as pd
import random
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
import re
remapear=list()
for file in sorted(listdir("./con_graphs/")):
    print("Procesando el archivo:" + file)
    G=nx.read_adjlist("./con_graphs/"+file)
    results = np.array(list(map(int, G.nodes())))
    results = results[results > 65535]
    for item in results:
        if not item in remapear:
            remapear.append(item)
```

```python
remap_arr=np.array(remapear)
print("Los nodos a remapear son: " + str (remap_arr))
new_nodes=np.arange(65536,65536+remap_arr.size)
print("Los nuevos valores de nodos son: " + str (new_nodes))
df_remap=pd.DataFrame()
df_remap["old_node"]=remap_arr
df_remap["new_node"]=new_nodes
df_remap.head()
```

```python
df_remap.to_csv("./remap.txt", index=False, sep=" ")
```

```python
import pandas as pd
import random
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
import re
for file in sorted(listdir("./con_graphs/")):
    print("Procesando el archivo:" + file)
    G=nx.read_adjlist("./con_graphs/"+file)
    G=nx.relabel_nodes(G,df_remap.set_index('old_node')['new_node'].to_dict())
    df=pd.DataFrame(G.edges())
    df.to_csv("./con_as_ren_graphs/"+file, index=False, header=False, sep=" ")
```

## Generación de gráficos y guardado de logs

A partir de las componentes conexas de mayor tamaño que se obtuvieron, se procesarán con el LaNet-vi cada uno de los grafos en orden cronológico.
Como algunos procesamientos falla, se incluye una variable "from_file" en el caso de que se quiera comenzar desde algún momento en particular.

```python
import pandas
import random
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
import subprocess
import os
def generar_graficos(flag=0, from_file="20140900.net"):
    #flag = 1 #1 si quiero procesar todo 0, si quiero procesar a partir del archivo indicado en "from_file".
    #from_file="20030200.net"
    for file in sorted(listdir("./con_as_ren_graphs/")):
        if(file==from_file):
            flag=1
        if(flag==1):
            f = open(file.split(".")[0]+".log", "w")
            print(file)
            subprocess.call(["../LaNet-vi_3.0.1/lanet","-input", "./con_as_ren_graphs/" + file, "-W", "1024", "-H", "768", "-render", "svg", "-decomp","kdenses","-names", "../LaNet-vi_3.0.1/asn.txt", "-edges", "1.0", "-bckgnd","white", "-logfile"], stdout=f)
            os.makedirs("./LaNet_logs_kdense/"+file.split(".")[0])
            os.rename("./log/coordinates.log", "./LaNet_logs_kdense/"+file.split(".")[0]+"/coordinates.log")
            os.rename("./log/cores.log", "./LaNet_logs_kdense/"+file.split(".")[0]+"/cores.log")
            os.rename("./log/pixels.log", "./LaNet_logs_kdense/"+file.split(".")[0]+"/pixels.log")
            os.rename("./"+file.split(".")[0]+".log", "./LaNet_logs_kdense/" + file.split(".")[0]+ "/" +file.split(".")[0]+".log")
            os.rename("./"+file.split(".")[0]+"_col_w_1024x768_names.svg", "../../../../Web/Topologías de Internet/ARK+BGP/svg_kdense/"+file.split(".")[0]+"_col_w_1024x768_names.svg")
            os.rename("./"+file.split(".")[0]+"_col_w_1024x768_namesSVG.png", "../../../../Web/Topologías de Internet/ARK+BGP/png_kdense/"+file.split(".")[0]+"_col_w_1024x768_namesSVG.png")
            f.close()
            f = open(file.split(".")[0]+".log", "w")
            print(file)
            subprocess.call(["../LaNet-vi_3.0.1/lanet","-input", "./con_as_ren_graphs/" + file, "-W", "1024", "-H", "768", "-render", "svg","-names", "../LaNet-vi_3.0.1/asn.txt", "-edges", "1.0", "-bckgnd","white", "-logfile"], stdout=f)
            os.makedirs("./LaNet_logs_kcores/"+file.split(".")[0])
            os.rename("./log/coordinates.log", "./LaNet_logs_kcores/"+file.split(".")[0]+"/coordinates.log")
            os.rename("./log/cores.log", "./LaNet_logs_kcores/"+file.split(".")[0]+"/cores.log")
            os.rename("./log/pixels.log", "./LaNet_logs_kcores/"+file.split(".")[0]+"/pixels.log")
            os.rename("./"+file.split(".")[0]+".log", "./LaNet_logs_kcores/" + file.split(".")[0]+ "/" +file.split(".")[0]+".log")
            os.rename("./"+file.split(".")[0]+"_col_w_1024x768_names.svg", "../../../../Web/Topologías de Internet/ARK+BGP/svg_kcores/"+file.split(".")[0]+"_col_w_1024x768_names.svg")
            os.rename("./"+file.split(".")[0]+"_col_w_1024x768_namesSVG.png", "../../../../Web/Topologías de Internet/ARK+BGP/png_kcores/"+file.split(".")[0]+"_col_w_1024x768_namesSVG.png")
            f.close()
```
